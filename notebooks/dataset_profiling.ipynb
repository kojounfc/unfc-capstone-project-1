All schema information captured in this notebook will be derived directly from:

`bigquery-public-data.thelook_ecommerce.INFORMATION_SCHEMA`



## Discovering Available Tables

The following query retrieves the list of all tables available in the dataset directly from BigQuery metadata.

```sql
SELECT table_name
FROM `bigquery-public-data.thelook_ecommerce.INFORMATION_SCHEMA.TABLES`
ORDER BY table_name;


## Discovering Schema (Columns and Data Types)

After identifying relevant tables, the following query retrieves column names and data types directly from BigQuery metadata.

```sql
SELECT
  table_name,
  column_name,
  data_type
FROM `bigquery-public-data.thelook_ecommerce.INFORMATION_SCHEMA.COLUMNS`
ORDER BY table_name, ordinal_position;


## Table Sizes (Row Counts)

To understand dataset scale, the following query provides row counts for each table:

```sql
SELECT
  table_name,
  row_count
FROM `bigquery-public-data.thelook_ecommerce.__TABLES__`
ORDER BY row_count DESC;

````python
import pandas as pd

# Load raw datasets
orders = pd.read_csv("../data/raw/orders.csv")
order_items = pd.read_csv("../data/raw/order_items.csv")
products = pd.read_csv("../data/raw/products.csv")
users = pd.read_csv("../data/raw/users.csv")

datasets = {
    "orders": orders,
    "order_items": order_items,
    "products": products,
    "users": users
}

# Profiling output
for name, df in datasets.items():
    print(f"\n=== {name.upper()} ===")
    print("Rows:", df.shape[0])
    print("Columns:", df.shape[1])
    print("\nColumn names:")
    print(df.columns.tolist())
    print("\nData types:")
    print(df.dtypes)
````
