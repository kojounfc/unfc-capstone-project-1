## Discovering Available Tables

The following query retrieves the list of all tables available in the dataset directly from BigQuery metadata.

```sql
SELECT table_name
FROM `bigquery-public-data.thelook_ecommerce.INFORMATION_SCHEMA.TABLES`
ORDER BY table_name;


## Discovering Schema (Columns and Data Types)

After identifying relevant tables, the following query retrieves column names and data types directly from BigQuery metadata.

```sql
SELECT
  table_name,
  column_name,
  data_type
FROM `bigquery-public-data.thelook_ecommerce.INFORMATION_SCHEMA.COLUMNS`
ORDER BY table_name, ordinal_position;


## Table Sizes (Row Counts)

To understand dataset scale, the following query provides row counts for each table:

```sql
SELECT
  table_name,
  row_count
FROM `bigquery-public-data.thelook_ecommerce.__TABLES__`
ORDER BY row_count DESC;


## Structural Observations (To be completed)

After running the above queries, this section will summarize key structural insights such as:

- Which tables are most relevant to transaction-level analysis  
- Which tables contain customer, product, and event-level data  
- Key identifier fields (e.g., primary keys and join fields)  
- Any notable schema constraints or limitations  

This section will be completed after reviewing query results.


## Captured Results (To be filled after running queries)

Paste or summarize the results from BigQuery here:

- **Tables discovered:** (paste list or key tables)
- **Row counts:** (paste top tables + counts)
- **Key join fields identified:** (e.g., order_id, user_id, product_id)
- **Notable fields for returns/profit erosion:** (e.g., status, returned_at, sale_price, cost)

This section provides evidence that schema profiling was performed using the authoritative source.
